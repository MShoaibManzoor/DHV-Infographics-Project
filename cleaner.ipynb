```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./EO108_18122023160120750.csv')
df
```

```python
# Checking data characteristics.
characteristics = pd.concat([df.isnull().sum(), df.nunique(), df.dtypes],
                        keys=['Null Values', 'Unique Values', 'Data Types'],
                        axis=1)
characteristics
```

```python
dump_columns = ['LOCATION', 'Country', 'VARIABLE', 'FREQUENCY', 'Frequency', 'TIME', 'Flag Codes', 'Flags']
for column in dump_columns:
    if column in df.columns:
        df = df.drop(columns=[column])
    else:
        print("{} is not the Dataset!".format(column))
```

```python
df_immigration = pd.read_csv('./Immigration_reason_stats.csv')
```

```python
columns = df_immigration.columns[1:]
df_immigration = df_immigration.groupby('Time')[columns].mean().reset_index()
```

```python
df_immigration = df_immigration.round(0).astype(int)
df_immigration.to_csv('immigration_stats_cleaned.csv')
```
